```python
from sklearn.datasets import make_regression
import pandas as pd

X,y = make_regression(n_samples = 100, n_features = 1, noise = 5, random_state = 42)
```

This code snippet utilizes the make_regression function from the sklearn.datasets module to create a synthetic dataset for regression analysis. Let's break it down step by step:

from sklearn.datasets import make_regression: This line imports the make_regression function from the sklearn.datasets module. sklearn is a popular machine learning library in Python, and make_regression is a utility function that generates a synthetic dataset suitable for regression analysis.
import pandas as pd: This line imports the pandas library with the alias pd. Pandas is another Python library commonly used for data manipulation and analysis. It provides data structures and functions to work with structured data, such as data frames.
X, y = make_regression(n_samples=100, n_features=1, noise=5, random_state=42): This line generates the synthetic dataset. Here's what each parameter means:
n_samples: Specifies the number of samples (data points) in the dataset. In this case, it generates 100 samples.
n_features: Specifies the number of features (independent variables) in the dataset. Here, it generates 1 feature.
noise: Controls the amount of random noise added to the target variable. Higher noise values result in more scattered data points around the regression line. In this case, the noise level is set to 5.
random_state: Sets the seed for the random number generator. This ensures reproducibility of the results. By setting it to 42, the same dataset will be generated every time this code is run.
X and y: These variables store the features (X) and the target variable (y) of the generated dataset, respectively. X is a two-dimensional array (100 rows by 1 column) containing the independent variable(s), and y is a one-dimensional array containing the dependent variable.

```python
data = pd.DataFrame({'X' : X.flatten(), 'y':y})
data
```

```python
data.info()
```

```python
window_size = 2

rolling_window = data['y'].rolling(window = window_size)
rolling_window
```

```python
rolling_window.mean()
```

```python
data['Rolling mean'] = rolling_window.mean()
data
```

```python
data['Rolling_sum_y'] = rolling_window.sum()
data
```

```python
data['Rolling_min_y'] = rolling_window.min()
data
```

```python
import numpy as np

np.random.choice(['A','B'], size = 100)
```

```python
data['Group'] = np.random.choice(['A','B'], size = 100)
data
```

```python
window_size = 3
group_data = data.groupby('Group')['y'].rolling(window = window_size).sum()
group_data
```

```python
def custom_mean(window):
  return window.mean(skipna = True)

window_size = 5

custom_window = data['y'].rolling(window = window_size)
data['custom_rolling_mean'] = custom_window.apply(custom_mean)
data
```
