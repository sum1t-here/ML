````python
new_df = df.dropna(thresh=3)
new_df
````

If you apply dropna with thresh=3 on the given DataFrame, it will drop rows where at least 3 non-NA values are not present.

In this case, the row with index 0 will be dropped because it has only one non-NA value ('Age'), while the other columns have NaN or None values.

````python
fillDf = df.fillna(value=0)
fillDf
````

## Filling missing value with mean value

````python
df['Age'].fillna(value = df['Age'].mean())
````
